```python
list=[]
N=50
Pmax=200
list.append(["P","bound","C","diff","reldiff"])
for P in range(1,Pmax):
    if P<=N:
        C=bound=2**P

    else:
        bound=(math.e*P/N)**N
        C=0
        for i in range(0,N-1): #summation
            C+=math.comb(P-1,i)
        C*=2
    
    diff=bound-C #space between bound and actual value#
    reldiff=diff/C #this size of the "error" relative to the actual value of C
    
    #change to scientific format
    if C>10000:
        bound = "{:.2e}".format(bound)
        C = "{:.2e}".format(C)
        diff = "{:.2e}".format(diff)
        reldiff = "{:.2e}".format(reldiff) 
    
    list.append([P,bound,C,diff,reldiff])

print(tabulate(list))
```
